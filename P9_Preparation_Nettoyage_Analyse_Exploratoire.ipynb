### <span style="color:#87cefa">Jointure</span>

```python
dispo_pop = pd.merge(dispo_vol, pop1, left_on = "Pays", right_on = "Pays", how = "outer", indicator = True)
dispo_pop
```

```python
# Affichage des valeurs Null
dispo_pop[dispo_pop.isnull().any(axis = 1)]
```

```python
# Suppression du Soudan dont les informations (population) ne pourront pas permetre une analyse précise
# Suppression des valeurs null restante car elle ne fournisse pas de données concernant l'importation et la disponibilité alimentaire
dispo_pop.dropna(axis = 0, inplace = True)
dispo_pop
```

```python
# Vérification des valeurs null
dispo_pop.isnull().sum()
```

```python
# Suppression de la colonne "_merge"
dispo_pop = dispo_pop.drop("_merge", axis = 1)
dispo_pop
```

```python
# Passage de la population en int
dispo_pop["pop2007"] = dispo_pop["pop2007"].astype(int)
dispo_pop["pop2017"] = dispo_pop["pop2017"].astype(int)
dispo_pop
```

```python
# Calcul de l'évolution de la population entre 2007 et 2017
dispo_pop["Tx evo_pop"] = round(((dispo_pop["pop2017"] / dispo_pop["pop2007"]) ** (1 / (2017 - 2007))) - 1, 2)

# Vérification des données populations avec la méthode TCAM
#dispo_pop["Tx evo_pop_vérifcalcul"] = round(dispo_pop["pop2007"] * ((1 + dispo_pop["TCAM"]) ** (2017 - 2007)), 2).astype(int)

dispo_pop
```

```python
# Calcul du taux d'importation
dispo_pop["Tx import"] = round((dispo_pop["Importation"] / dispo_pop["Disponibilite interieur"]), 2)
dispo_pop
```

```python
# Calcul du taux de production
dispo_pop["Tx prod"] = round((dispo_pop["Production"] / dispo_pop["Disponibilite interieur"]), 2)
dispo_pop
```

```python
# Liste des pays présent dans les données final
liste_pays = dispo_pop['Pays'].tolist()
```

```python
# Carte centrée sur les coordonnées
map_world = folium.Map(location=[0, 0], zoom_start=2)

# Tuiles (tiles layer) pour afficher les fonds de carte
folium.TileLayer('cartodbpositron').add_to(map_world)

# Géocodeur pour obtenir les coordonnées géographiques
geolocator = Nominatim(user_agent="my_app")

# Liste avec marqueurs pour chaque pays
for country in liste_pays:
    # Coordonnées géographiques du pays
    location = geolocator.geocode(country)
    
    if location:
        # Coordonnées géographiques pour positionner le marqueur
        latitude = location.latitude
        longitude = location.longitude
        
        # Marqueur pour chaque pays
        folium.Marker(location=[latitude, longitude], popup=country).add_to(map_world)

# Afficher la carte
map_world
```
